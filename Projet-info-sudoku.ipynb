```python
import tkinter as tk  # Importation du module principal pour l'interface graphique
from tkinter import messagebox  # Importation spécifique pour les fenêtres d'alerte (Pop-up)

# Grille de départ : 0 représente une case vide que l'étudiant doit remplir
grille_jeu = [
    [0, 0, 0, 2, 6, 0, 7, 0, 1],
    [6, 8, 0, 0, 7, 0, 0, 9, 0],
    [1, 9, 0, 0, 0, 4, 5, 0, 0],
    [8, 2, 0, 1, 0, 0, 0, 4, 0],
    [0, 0, 4, 6, 0, 2, 9, 0, 0],
    [0, 5, 0, 0, 0, 3, 0, 2, 8],
    [0, 0, 9, 3, 0, 0, 0, 7, 4],
    [0, 4, 0, 0, 5, 0, 0, 3, 6],
    [7, 0, 3, 0, 1, 8, 0, 0, 0]
]

# Grille contenant la solution correcte pour la comparaison
solution = [
    [4, 3, 5, 2, 6, 9, 7, 8, 1],
    [6, 8, 2, 5, 7, 1, 4, 9, 3],
    [1, 9, 7, 8, 3, 4, 5, 6, 2],
    [8, 2, 6, 1, 9, 5, 3, 4, 7],
    [3, 7, 4, 6, 8, 2, 9, 1, 5],
    [9, 5, 1, 7, 4, 3, 6, 2, 8],
    [5, 1, 9, 3, 2, 6, 8, 7, 4],
    [2, 4, 8, 9, 5, 7, 1, 3, 6],
    [7, 6, 3, 4, 1, 8, 2, 5, 9]
]

# Liste vide qui va stocker les objets 'Entry' pour pouvoir lire leurs valeurs plus tard
toutes_les_cases = []

# --- FONCTION DE VÉRIFICATION ---
def verifier_la_grille():
    nombre_erreurs = 0  # Compteur pour les fautes de frappe
    nombre_cases_vides = 0  # Compteur pour vérifier si tout est rempli
    
    for i in range(9):  # Boucle sur les indices des lignes (0 à 8)
        for j in range(9):  # Boucle sur les indices des colonnes (0 à 8)
            case_actuelle = toutes_les_cases[i][j]  # On récupère le widget Entry à cette position
            valeur_saisie = case_actuelle.get()  # .get() récupère le texte écrit dans la case
            
            if valeur_saisie == "":  # Si la chaîne de caractères est vide
                nombre_cases_vides = nombre_cases_vides + 1  # Incrémentation simple
                case_actuelle.config(bg="white")  # On remet le fond en blanc
            else:
                chiffre_joueur = int(valeur_saisie)  # Conversion du texte (str) en nombre (int)
                chiffre_reel = solution[i][j]  # Extraction de la solution correspondante
                
                if chiffre_joueur == chiffre_reel:  # Test d'égalité
                    if grille_jeu[i][j] == 0:  # On ne colorie que si c'était une case vide au début
                        case_actuelle.config(bg="#c2f0c2")  # Fond vert clair (code Hexadécimal)
                else:
                    case_actuelle.config(bg="#ffcccc")  # Fond rouge clair
                    nombre_erreurs = nombre_erreurs + 1  # On compte l'erreur
    
    # Affichage des résultats selon les compteurs
    if nombre_cases_vides > 0:
        messagebox.showwarning("Attention", "Grille incomplète !")
    elif nombre_erreurs > 0:
        messagebox.showerror("Erreur", "Certaines cases sont fausses.")
    else:
        messagebox.showinfo("Succès", "Félicitations, tout est juste !")

# --- CONSTRUCTION DE L'INTERFACE ---
fenetre = tk.Tk()  # Création de la fenêtre principale
fenetre.title("Sudoku Bio-Info")  # Titre de la fenêtre

for i in range(9):  # Création des lignes
    liste_ligne = []  # Liste temporaire pour stocker les cases d'une ligne
    for j in range(9):  # Création des colonnes
        # Création du widget Entry (Saisie de texte)
        case = tk.Entry(fenetre, width=2, font=('Arial', 20), justify='center', bd=1, relief="solid")
        
        # Calcul des marges pour créer l'effet de blocs 3x3
        marge_x = 1
        if j % 3 == 0: marge_x = 5  # Ajoute 5 pixels à gauche toutes les 3 colonnes
        
        marge_y = 1
        if i % 3 == 0: marge_y = 5  # Ajoute 5 pixels en haut toutes les 3 lignes
        
        # Placement de la case dans la grille (row=ligne, column=colonne)
        case.grid(row=i, column=j, padx=(marge_x, 1), pady=(marge_y, 1))
        
        valeur_initiale = grille_jeu[i][j]
        if valeur_initiale != 0:
            case.insert(0, str(valeur_initiale))  # Insère le chiffre au début (index 0)
            case.config(state='readonly', readonlybackground='#dfe6e9')  # Verrouille la case
            
        liste_ligne.append(case)  # Ajout de la case à la ligne en cours
    toutes_les_cases.append(liste_ligne)  # Ajout de la ligne à la liste globale

# Création du bouton de validation
bouton = tk.Button(fenetre, text="Vérifier la solution", command=verifier_la_grille)
bouton.grid(row=9, column=0, columnspan=9, pady=20)  # Placement en bas de la grille

fenetre.mainloop()  # Lancement de la boucle infinie pour afficher la fenêtre
```
